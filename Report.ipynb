# Visualizing CNN classifier #

## Introduction ##

The assignment aims to investigate interpretability in a neural network using Class Activation Mapping (CAM).

# Method #

For this image classificator, Resnet18 (pretrained) is utilized. It has pre-trained weights for the ImageNet dataset, containing 1000 classes.

Torch-CAM is utalised in a similar way as in the course repo with a number of modifications. Three classes have been picked with one positive picture containing an object with the class and one picture without the class pressent.

The negative tests have been performed with pictures featuring objects not pressent in the ImageNet dataset. As a result, the model associates visually similar features to the closest known class.

# Result #

 ### Test 1a: Chimpanzee, positive example  ###

```python
from ImageClassificator import ImageClassificator

image_path = "images/Chimpanzee.jpg"
target_class = 367 # Chimpanzee
ImageClassificator(image_path, target_class)
```

 ### Test 1b: Chimpanzee, negative example  ###

```python
image_path = "images/Yoda.jpg"
target_class = 367 # Chimpanzee
ImageClassificator(image_path, target_class)
```

In test 1a, the model predicts chimpanzee with a confidence of 0.884823. The activation map highlights the face and the shoulder.

In test 1b, When forcing the chimpanzee class, the activation map highlights the ears and facial structure, indicating that the model detects some primate-like features despite the incorrect classification.

### Test 2a: Pillow, positive example ###

```python
image_path = "images/Pillow.jpg"
target_class = 721  # Pillow
ImageClassificator(image_path, target_class)
```

### Test 2b: Pillow, negative example ###

```python
image_path = "images/ettan_original_prilla.webp"
target_class = 721  # Pillow
ImageClassificator(image_path, target_class)
```

In test 2a, the model predicts pillow with a confidence of 0.506205. The activation map highlights the contours of the pillow on the short sides, indicating that the model focuses on the shape.

In test 2b, When forcing the pillow class on a portion of Ettan snus, the activation map highlights the center and the short sides, indicating that the model detects the pillow resembling shape and surface of a portion of snus and thus makes incorrect classification.

### Test 3a: Parachute, positive example ###

```python
image_path = "images/parachute.jpg"
target_class = 701   # Parachute
ImageClassificator(image_path, target_class)
```

### Test 3b: Parachute, negative example ###
